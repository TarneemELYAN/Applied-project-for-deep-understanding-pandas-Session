## Filtering Columns & rows

using uniqe method

```python
# Assuming your DataFrame is named 'df'
unique_bedroom_counts = df['bedrooms'].unique()
print(unique_bedroom_counts)
#unique_bedroom_counts will contain an array of unique bedroom counts present in the 'bedrooms' column.
```

Filtering Rows Based on Conditions:
 filter rows in a DataFrame based on certain conditions using boolean indexing.

```python
# Filtering rows where 'bedrooms' is greater than or equal to 3
filtered_rows = df[df['bedrooms'] >= 3]
print(filtered_rows)
# Combining multiple conditions using logical operators (& for AND, | for OR)
filtered_rows1 = df[(df['bedrooms'] >= 3) & (df['bathrooms'] >= 2)]
print(filtered_rows1)
```

## Data Selection and Indexing:

Using Square Brackets [] Indexing

```python
# Selecting specific columns using square brackets
selected_columns = df[['id', 'date', 'price']]
selected_columns
```

Using .loc[] and .iloc[]:

```python
#loc[] is label-based indexing
# Using .loc[] for label-based indexing
selected_data1 = df.loc[:, ['id', 'date', 'price']]
print(selected_data1)
#iloc[] is integer-based indexing
# Using .iloc[] for integer-based indexing
selected_data2 = df.iloc[:, [0, 1, 2]]
print(selected_data2)
```

Grouping and Aggregating Data:

```python
# Grouping by 'bedrooms' and calculating mean price and max sqft_living
grouped_data = df.groupby('bedrooms').agg({'price': 'mean', 'sqft_living': 'max'})
grouped_data
```

## Correlation Analysis

```python
df.corr()
```

```python
# Filter columns for corr calculation
columns_of_interest = ['price', 'bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'waterfront', 'view', 'condition', 'grade', 'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated', 'zipcode', 'lat', 'long', 'sqft_living15', 'sqft_lot15']
correlations = df[columns_of_interest].corr()
# you print the correlations of each column with the 'price' column, sorted in descending order.

print(correlations['price'].sort_values(ascending=False))
```
